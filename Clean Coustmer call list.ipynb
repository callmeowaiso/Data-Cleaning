Cleaning Coustmer call list

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Data Cleaning\call list.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df= df.drop(columns="Not_Useful_Column")
df
```

```python
df['Last_Name'] = df["Last_Name"].str.lstrip("123./_")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df['Paying Customer'] = df['Paying Customer'].str.replace('N/a','')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')

df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace= True)

df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace= True)

df
```

```python
df = df.reset_index(drop=True)
df
```

```python
# Export to Excel in the same folder
df.to_excel('./cleaned_data.xlsx', index=False)
```

Visualizing the dataset

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count the occurrences of 'Y' and 'N'
contact_counts = df['Do_Not_Contact'].value_counts()
print(contact_counts)
```

```python
# Create a pie chart
plt.figure(figsize=(3, 3))
plt.pie(contact_counts, labels=contact_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Do Not Contact Preferences')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
